```python
import pandas as pd
import numpy as np
from BLP import UtilityEstimation
```

```python
theta_1_true = np.array([-0.5, 1.0, 0.8])  
pi_true = np.array([[0.1, -0.05],
                    [0.2, 0.1],
                    [-0.1, 0.05]])      
sigma_true = np.array([0.2, 0.5, 0.3]) 
```

```python
df_char=pd.read_csv('characteristics.csv')
df_char.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df_char.to_numpy().shape
```

```python
df_demo=pd.read_csv('demographics.csv')
df_demo.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df_demo.to_numpy().shape
```

```python
df_IV=pd.read_csv('IV.csv')
df_IV.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df_IV.to_numpy().shape
```

```python
df_share=pd.read_csv('share.csv')
```

```python
df_share['Share'].to_numpy().shape
```

```python
obj=UtilityEstimation(
    X=df_char.to_numpy(),s_jt=df_share['Share'].to_numpy(), D=df_demo.to_numpy(), ns=1000, seed=42
)
```

```python
delta,status=obj.blp_contraction_mapping(theta_1=theta_1_true, pi=pi_true, sigma=sigma_true)
```

```python
status
```

```python
pd.DataFrame(delta).head()
```

```python
error =obj.compute_xi(delta=delta,theta_1=theta_1_true,X=df_char.to_numpy())
```

```python
error
```

```python
m=obj.compute_moments(xi=error,Z=df_IV.to_numpy())
```

```python
m
```

```python
pi_true.flatten()
```

```python
objective=obj.gmm_objective(
    theta_params=np.hstack([theta_1_true,pi_true.flatten(),sigma_true]),
    Z=df_IV.to_numpy()
)
```

```python
objective
```

```python
dict_r=obj.estimate_parameters(
    initial_theta=np.hstack([theta_1_true,pi_true.flatten(),sigma_true]),
    Z=df_IV.to_numpy()
)
```
